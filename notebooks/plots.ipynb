```python
from matplotlib import pyplot as plt
import pandas as pd

plt.plot(range(0, 17), [1500] * 17)


def plot(folder):
    df = pd.read_csv(f'{folder}/logs.csv')
    ret = df['mean_episode_return'].ewm(alpha=0.1).mean()
    plt.plot(df['train_seconds'] / 3600, ret, label=folder)
    # plt.plot(df['train_seconds'] / 3600, df['step'], label=folder)

plot('../latest')
plt.legend(loc='lower right')
plt.show()
```
