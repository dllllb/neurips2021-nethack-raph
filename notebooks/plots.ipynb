```python
from matplotlib import pyplot as plt
import pandas as pd

x = 250
plt.plot(range(0, x), [1500] * x)


def plot(folder):
    df = pd.read_csv(f'{folder}/logs.csv')
    ret = df['mean_episode_return'].ewm(alpha=0.1).mean()
    plt.plot(df['step'] / 1e5, ret, label=folder)
    # plt.plot(df['train_seconds'] / 3600, ret, label=folder)

plot('../latest')                                 # previous master net
plot('../nethack_baselines/torchbeast/latest')    # latest net
plt.legend(loc='lower right')
plt.show()
```
